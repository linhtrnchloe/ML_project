```python
# Wind speed (resulting speed from U and V components)
ancona['wind_speed'] = np.sqrt(ancona['Wind-Speed (U)']**2 + ancona['Wind-Speed (V)']**2)

# Wind direction in radians
ancona['wind_dir_rad'] = np.arctan2(ancona['Wind-Speed (V)'], ancona['Wind-Speed (U)'])

# Convert to degrees from 0 to 360 (optional but nicer)
ancona['wind_dir_deg'] = (np.degrees(ancona['wind_dir_rad']) + 360) % 360
```

```python
# Basic calendar features
ancona['hour'] = ancona['Date'].dt.hour
ancona['dayofweek'] = ancona['Date'].dt.dayofweek  # Monday=0, Sunday=6
ancona['month'] = ancona['Date'].dt.month
ancona['is_weekend'] = ancona['dayofweek'] >= 5
```

```python
def month_to_season(m):
    if m in [12, 1, 2]:
        return 0  # winter
    elif m in [3, 4, 5]:
        return 1  # spring
    elif m in [6, 7, 8]:
        return 2  # summer
    else:
        return 3  # autumn

ancona['season'] = ancona['month'].apply(month_to_season)
```

```python
ancona['hour_sin'] = np.sin(2 * np.pi * ancona['hour'] / 24)
ancona['hour_cos'] = np.cos(2 * np.pi * ancona['hour'] / 24)
```

```python
max_lag = 24  # use last 24 hours

for lag in range(1, max_lag + 1):
    ancona[f'PM25_lag_{lag}'] = ancona.groupby('station_name')['PM2.5'].shift(lag)
```

```python
results = []

for st in stations:
    # Filter data for this station
    train_st = train_data[train_data['station_name'] == st]
    val_st   = val_data[val_data['station_name'] == st]
    test_st  = test_data[test_data['station_name'] == st]
    
    # In case some station has no data in a split, skip it
    if len(train_st) == 0 or len(val_st) == 0 or len(test_st) == 0:
        continue
    
    X_train_st = train_st[feature_cols]
    y_train_st = train_st['PM2.5']
    
    X_val_st = val_st[feature_cols]
    y_val_st = val_st['PM2.5']
    
    X_test_st = test_st[feature_cols]
    y_test_st = test_st['PM2.5']
    
    # Create and fit station-specific RF
    rf_local = RandomForestRegressor(
        **best_params,
        random_state=42,
        n_jobs=-1
    )
    rf_local.fit(X_train_st, y_train_st)
    
    # Predictions
    y_val_pred_st  = rf_local.predict(X_val_st)
    y_test_pred_st = rf_local.predict(X_test_st)
    
    # Metrics
    mae_val  = mean_absolute_error(y_val_st, y_val_pred_st)
    rmse_val = np.sqrt(mean_squared_error(y_val_st, y_val_pred_st))
    
    mae_test  = mean_absolute_error(y_test_st, y_test_pred_st)
    rmse_test = np.sqrt(mean_squared_error(y_test_st, y_test_pred_st))
    
    results.append({
        'station': st,
        'val_mae': mae_val,
        'val_rmse': rmse_val,
        'test_mae': mae_test,
        'test_rmse': rmse_test
    })
    
    print(f"Station: {st}")
    print(f"  Val  MAE={mae_val:.3f}, RMSE={rmse_val:.3f}")
    print(f"  Test MAE={mae_test:.3f}, RMSE={rmse_test:.3f}")
    print("-" * 40)

results_df = pd.DataFrame(results)
results_df
```

We use pipelines to ensure consistent preprocessing (imputation + scaling where needed) and evaluate using **MAE, RMSE, and R²**.


```python
import numpy as np
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit, GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR

# Safety check: confirm we have the key matrices from previous cells
for name in ['X_train','y_train','X_val','y_val','X_test','y_test','feature_cols']:
    if name not in globals():
        raise NameError(f"Missing variable '{name}'. Run the earlier cells that create the train/val/test split first.")

# Helper: metric function
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def regression_metrics(y_true, y_pred):
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))  # <-- works in all sklearn versions
    r2 = r2_score(y_true, y_pred)
    return {'MAE': mae, 'RMSE': rmse, 'R2': r2}


# Time-series CV on the TRAIN set only (respects ordering)
tscv = TimeSeriesSplit(n_splits=5)
```

```python
# ---- Define model pipelines + hyperparameter grids ----

# NOTE:
# - Scaling is important for kNN and SVR.
# - Linear regression benefits from scaling only for numerical stability (not required),
#   but we keep the preprocessing consistent.

preprocess_scaled = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

preprocess_unscaled = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median'))
])

models = {
    'LinearRegression': (
        Pipeline(steps=[
            ('prep', preprocess_scaled),
            ('model', LinearRegression())
        ]),
        {}  # no hyperparameters to tune for basic OLS
    ),
    'kNN': (
        Pipeline(steps=[
            ('prep', preprocess_scaled),
            ('model', KNeighborsRegressor())
        ]),
        {
            'model__n_neighbors': [3, 5, 7, 11, 15, 21],
            'model__weights': ['uniform', 'distance'],
            'model__p': [1, 2],  # 1=Manhattan, 2=Euclidean
        }
    ),
    'DecisionTree': (
        Pipeline(steps=[
            ('prep', preprocess_unscaled),
            ('model', DecisionTreeRegressor(random_state=42))
        ]),
        {
            'model__max_depth': [None, 5, 10, 20],
            'model__min_samples_split': [2, 10, 50],
            'model__min_samples_leaf': [1, 5, 20]
        }
    ),
    'RandomForest': (
        Pipeline(steps=[
            ('prep', preprocess_unscaled),
            ('model', RandomForestRegressor(random_state=42, n_jobs=1))
        ]),
        {
            'model__n_estimators': [200, 500],
            'model__max_depth': [None, 10, 20],
            'model__min_samples_leaf': [1, 5, 20],
            'model__max_features': ['sqrt', 0.6, 1.0]
        }
    ),
    'SVR': (
        Pipeline(steps=[
            ('prep', preprocess_scaled),
            ('model', SVR())
        ]),
        {
            'model__kernel': ['rbf', 'linear'],
            'model__C': [1, 10, 50],
            'model__epsilon': [0.05, 0.1, 0.2],
            'model__gamma': ['scale', 'auto']
        }
    )
}
```

```python
# ---- Train, tune (if grid given), and evaluate ----

results = []
best_estimators = {}

for model_name, (pipe, grid) in models.items():
    print(f"\n=== {model_name} ===")

    if grid:
        # Use negative RMSE as CV objective
        gscv = GridSearchCV(
            estimator=pipe,
            param_grid=grid,
            scoring='neg_root_mean_squared_error',
            cv=tscv,
            n_jobs=1,
            refit=True
        )
        gscv.fit(X_train, y_train)
        best_model = gscv.best_estimator_
        best_params = gscv.best_params_
        best_cv_rmse = -gscv.best_score_
        print("Best params:", best_params)
        print("Best CV RMSE:", round(best_cv_rmse, 4))
    else:
        best_model = pipe.fit(X_train, y_train)
        best_params = {}

    best_estimators[model_name] = best_model

    # Validation
    y_val_pred = best_model.predict(X_val)
    val_m = regression_metrics(y_val, y_val_pred)

    # Test
        'Val_RMSE': val_m['RMSE'],
        'Val_R2': val_m['R2'],
        'Test_MAE': test_m['MAE'],
        'Test_RMSE': test_m['RMSE'],
        'Test_R2': test_m['R2'],
        'Best_Params': best_params
    })

results_df = pd.DataFrame(results).sort_values('Test_RMSE')
results_df
```

```python
# Quick readable summary (sorted by Test RMSE)
display_cols = ['Model','Val_RMSE','Val_MAE','Val_R2','Test_RMSE','Test_MAE','Test_R2']
results_df[display_cols].reset_index(drop=True)
```

```python
# Optional: feature importance for tree-based models (helps in the project report)
# Works for DecisionTree and RandomForest because they expose feature_importances_

def get_feature_importance(estimator, feature_names):
    model = estimator.named_steps['model']
    if hasattr(model, 'feature_importances_'):
        imp = pd.Series(model.feature_importances_, index=feature_names).sort_values(ascending=False)
        return imp
    return None

for name in ['DecisionTree', 'RandomForest']:
    if name in best_estimators:
        imp = get_feature_importance(best_estimators[name], feature_cols)
        if imp is not None:
            print(f"\nTop 15 features for {name}:")
            display(imp.head(15).to_frame('importance'))
```
